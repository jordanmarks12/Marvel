```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.businessinsider.com/marvel-movies-ranked-how-much-money-at-global-box-office-2021-11"
html = requests.get(url)
bsobj = BeautifulSoup(html.content, "html.parser")
```

```python
print(html)
```

```python
df = pd.DataFrame()
```

```python
# Movie Titles + Year
```

```python
movies = bsobj.find_all('div', class_='slide keyline first-slide')
#print(movies)
movie_titles = []
worldwide_boxoffice = []
total_domestic_boxoffice = []
domestic_opening_weekend = []
budget = []
for movie in movies:
    title = movie.find('h2', class_='slide-title-text').text
    movie_titles.append(title)
    
    money_worldwide = movie.find('p')
    #print(money_worldwide.text)
    money_domestic = money_worldwide.next_sibling
    #print(money_domestic.text)
    money_domestic_opening = money_domestic.next_sibling
    #print(money_domestic_opening.text)
    money_budget = money_domestic_opening.next_sibling
    #print(money_budget.text)
    
worldwide_boxoffice.append(money_worldwide.text)
#print(worldwide_boxoffice)
total_domestic_boxoffice.append(money_domestic.text)
#print(total_domestic_boxoffice)
domestic_opening_weekend.append(money_domestic_opening.text)
#print(domestic_opening_weekend)
budget.append(money_budget.text)
#print(budget)
    

movies_2 = bsobj.find_all('div', class_='slide keyline')
#print(movies_2)
for movie_2 in movies_2:
    title_2 = movie_2.find('h2', class_='slide-title-text').text
    movie_titles.append(title_2)
    
    money_worldwide_2 = movie_2.find('p')
    #print(money_worldwide_2.text)
    money_domestic_2 = money_worldwide_2.next_sibling
    #print(money_domestic_2.text)
    money_domestic_opening_2 = money_domestic_2.next_sibling
    #print(money_domestic_opening_2.text)
    money_budget_2 = money_domestic_opening_2.next_sibling
    #print(money_budget_2.text)
    
    worldwide_boxoffice.append(money_worldwide_2.text)
    total_domestic_boxoffice.append(money_domestic_2.text)
    domestic_opening_weekend.append(money_domestic_opening_2.text)
    budget.append(money_budget_2.text)

#print(worldwide_boxoffice)
#print(total_domestic_boxoffice)
#print(domestic_opening_weekend)
#print(budget)

#### budget isnt included for some(10 rings, wakanda forever, and released during corona virus is added to the budget section
    
    
movies_3 = bsobj.find_all('div', class_='slide keyline last-slide')
#print(movies_3)
for movie_3 in movies_3:
    title_3 = movie_3.find('h2', class_='slide-title-text').text
    movie_titles.append(title_3)

    money_worldwide_3 = movie_3.find('p')
    #print(money_worldwide_3.text)
    money_domestic_3 = money_worldwide_3.next_sibling
    #print(money_domestic_3.text)
    money_domestic_opening_3 = money_domestic_3.next_sibling
    #print(money_domestic_opening_3.text)
    money_budget_3 = money_domestic_opening_3.next_sibling
    #print(money_budget_3.text)
    
    worldwide_boxoffice.append(money_worldwide_3.text)
    total_domestic_boxoffice.append(money_domestic_3.text)
    domestic_opening_weekend.append(money_domestic_opening_3.text)
    budget.append(money_budget_3.text)


#print(worldwide_boxoffice)
#print(total_domestic_boxoffice)
print(domestic_opening_weekend)
#print(budget)
```

```python
# DataFrame
```

```python
cols = ['Titles', "Worldwide Box-Office", "Total Domestic Box-Office", "Domestic Opening Weekend Box-Office", "Budget"]
insider = pd.DataFrame(columns=cols)
insider['Titles'] = movie_titles
insider['Worldwide Box-Office'] = worldwide_boxoffice
insider['Total Domestic Box-Office'] = total_domestic_boxoffice
insider['Domestic Opening Weekend Box-Office'] = domestic_opening_weekend
insider['Budget'] = budget
insider
```
